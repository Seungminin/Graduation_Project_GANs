```python
from PIL import Image
import matplotlib.pyplot as plt
import random

image = Image.open("dog_cat.jpg")

# 16등분 나누기 4*4
rows = 4  # 행
cols = 4  # 열
tile_width = image.width // cols
tile_height = image.height // rows

tiles = []

for row in range(rows):
    for col in range(cols):
        left = col * tile_width
        upper = row * tile_height
        right = (col + 1) * tile_width if col < cols - 1 else image.width
        lower = (row + 1) * tile_height if row < rows - 1 else image.height
        box = (left, upper, right, lower)
        tiles.append(image.crop(box))

fig, axes = plt.subplots(rows, cols, figsize=(cols * 3, rows * 3))

for idx, tile in enumerate(tiles):
    row = idx // cols
    col = idx % cols
    axes[row, col].imshow(tile)
    axes[row, col].axis("off")  # 축 제거

plt.tight_layout()
plt.show()

'''# 타일 저장
main_object_tile_index = random.choice(range(len(tiles)))
tiles[main_object_tile_index].save("main_tile.png")
for idx, tile in enumerate(tiles):
    tile.save(f"tile_{idx}.png")'''
```
